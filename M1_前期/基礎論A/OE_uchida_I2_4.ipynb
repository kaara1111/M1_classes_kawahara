```python
import sympy
import random
```

```python
def exEuclid(a,b):
  x = sympy.gcdex(a,b)
  xa = [x[2],x[0],x[1]]
  return xa
```

```python
def inv(a,n):
  x = exEuclid(a,n)
  a_ = x[1]
  return a_%n
```

```python
def mod_binary(g,k,n):
  k2 = bin(k)[2:] #0bを排除した。k2はstr型。
  k2_1 = k2[:len(k2)-1] #2進数の一番最後の0,1は処理を別にするために排除した。
  x = 1
  g = g%n

  for i in k2_1:
    if(i == '1'):
      x = ((x*g)**2)%n
    else:
      x = (x**2)%n

  if(k2[len(k2)-1] == '1'):
    x = x*g

  return x%n
```

```python
def fermat_test(n,a):
  return (mod_binary(a,n-1,n) == 1)
```

```python
def fermat_primegen(iv,k):
  n = iv -1
  flag = 0
  while(flag == 0):
    n += 1
    flag = 1
    a = sympy.primerange(sympy.prime(k)+1)

    for i in a:
      if(fermat_test(n,i) != 1):
        flag = 0
        break

  return n
```

```python
def fermat_primegen2(iv_bit, k):
  min = 2**(iv_bit - 1)
  max = 2**iv_bit
  random_value = random.randrange(min,max)
  return fermat_primegen(random_value,k)
```

```python
def rsaKeygen(p,q):
  n = p*q
  z = sympy.lcm(p-1,q-1)
  x = []
  for i in range(1000):
    if(sympy.gcd(i,z) == 1):
      x.append(i)
  e = x[-1]
  d = inv(e,z)
  return n,e,d
```

```python
def rsaEnc(m,n,e):
  return mod_binary(m,e,n)
```

```python
def rsaDec(c,n,d):
  return mod_binary(c,d,n)
```
