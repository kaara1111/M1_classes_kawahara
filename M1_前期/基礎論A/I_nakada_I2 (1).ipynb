```python
#演習2.4, 石川高専, 中田万喜
#(1)

import random
import sympy

def mod_binary(g, k, n):     #nを法とするバイナリ法 (output: g^k(mod n))
    k_binary = []
    while k != 0:
        k_binary.append(k % 2)
        k = k // 2
        if k == 1:
            k_binary.append(k)
            k = 0
    # バイナリ法によるべき乗計算
    y = 1
    for i in reversed(range(len(k_binary))):
        if k_binary[i] == 1:
            y = (y * y % n) * g % n
        else:
            y = (y * y % n)    
    return y
a = 2

def fermat_test(n):      #a^(n-1) mod n が1だったらTrueを返す(output : True or False)
    if mod_binary(a, n - 1, n) != 1:
            return False
    return True


def PNG(n):     #fermat_testを満たす, nビットの乱数を生成(output : int)
    while True:
        p = random.randint(2**(n-1),2**n-1)
        if fermat_test(p) == True:
            return p

p = PNG(512)
q = PNG(512) #p, qは自分で保存
n = p * q
e = 65537
print('n = ')
print(n)
print('e = ')
print(e)
```
