```python
# 演習1-1(コードを追加してプログラムを完成させよ)

import time, sympy, numpy # 必要なモジュールをインポート

def statistics(data): # 最小値、最大値、平均、分散の計算
    return [min(data), max(data), numpy.mean(data), numpy.var(data)]

def time_check(ts, te): # 実行時間の計測
    return te - ts
    
a = 778851447666082307730996709
b = 1136041827995209856154025231
c = 639745742161865592519176783

Ta = [] # aの因数分解の時間を100回分格納するリスト
Tb = [] # bの因数分解の時間を100回分格納するリスト
Tc = [] # cの因数分解の時間を100回分格納するリスト

for i in range(100): #a, b, c の因数分解を100回実行し、各回の因数分解の時間をTa, Tb, Tcに格納
    ts = time.time()# 現時点の時刻を取得
    fact_a = sympy.factorint(a)# aの因数分解の結果を取得
    te = time.time()# 因数分解1回後の時刻を取得
    # Taに因数分解1回分の実行時間を格納
    Ta.append(te - ts)
    
    # aの因数分解の場合を参考に、b, cの因数分解の結果と実行時間を100回分取得するコードを追加
    ts = time.time()# 現時点の時刻を取得
    fact_b = sympy.factorint(b)# aの因数分解の結果を取得
    te = time.time()# 因数分解1回後の時刻を取得
    # Taに因数分解1回分の実行時間を格納
    Tb.append(te - ts)

    ts = time.time()# 現時点の時刻を取得
    fact_c = sympy.factorint(c)# aの因数分解の結果を取得
    te = time.time()# 因数分解1回後の時刻を取得
    # Taに因数分解1回分の実行時間を格納
    Tc.append(te - ts)

result_a = [min(Ta), max(Ta), numpy.mean(Ta), numpy.var(Ta)] # Ta を用いてaの因数分解の計算時間の最小値、最大値、平均、分散を取得
result_b = [min(Tb), max(Tb), numpy.mean(Tb), numpy.var(Tb)]# Tb を用いてbの因数分解の計算時間の最小値、最大値、平均、分散を取得
result_c = [min(Tc), max(Tc), numpy.mean(Tc), numpy.var(Tc)]# Tc を用いてcの因数分解の計算時間の最小値、最大値、平均、分散を取得

print("aの素因数分解の計算時間(秒)の最小値={}, 最大値={}, 平均={}, 分散={}".format(result_a[0], result_a[1], result_a[2], result_a[3]))
print("bの素因数分解の計算時間(秒)の最小値={}, 最大値={}, 平均={}, 分散={}".format(result_b[0], result_b[1], result_b[2], result_b[3]))
print("cの素因数分解の計算時間(秒)の最小値={}, 最大値={}, 平均={}, 分散={}".format(result_c[0], result_c[1], result_c[2], result_c[3]))
print("aの素因数分解結果：", fact_a)
print("bの素因数分解結果：", fact_b)
print("cの素因数分解結果：", fact_c)

# a, b, cの因数分解の実行時間について考察を記入しよう
# aは比較的素因数の大きさが小さいものが多いため，比較的早く終わり，bは素因数の大きさが大きいものが多いため，aよりも時間がかかる．cはaとbの中間の時間がかかる．
# aの分散が大きいのは，因数分解にかかる平均的な時間が小さいため，少しの誤差でも大きく分散が広がってしまうためである．
# bの分散がcと比べて大きいのは，素因数の探索の初期値によって，計算時間が大きく変わってしまうためであると考えられる．
```

```python
# 演習2(コードを追加してプログラムを完成させよ)

from scipy.stats import chisquare
from secrets import choice
from random import randint
from collections import Counter

alpha = 0.01 # 有意水準

# secrets.choiceによるさいころ投げ
def dice1(n):
    T = [] # さいころを投げた結果を格納
    D = [1,2,3,4,5,6] # さいころの目1～6
    # n回さいころを振って出た目をTに格納するコードを追加
    for i in range(n):
        T.append(choice(D))

    return T

# random.randintによるさいころ投げ
def dice2(n):
    # dice1(n)を参考にコードを追加し、関数を完成させよ
    T = []
    for i in range(n):
        T.append(randint(1,6))
    
    return T

count_non_random1 = 0
count_non_random2 = 0

for i in range(100):
    T1 = dice1(60000) # 60000回サイコロを振った結果をT1に格納（secret.choice利用)
    T2 = dice2(60000) # 60000回サイコロを振った結果をT2に格納（random.randint利用)
    c1 = Counter(T1) # c1にT1の中にさいころの各目が何個入っているか格納. c1[i]でさいころの目iが何個入っているか確認できる.
    c2 = Counter(T2) # c2にT2の中にさいころの各目が何個入っているか格納. c2[i]でさいころの目iが何個入っているか確認できる.
    
    chi1 = [c1[1], c1[2], c1[3], c1[4], c1[5], c1[6]] # dice1に関する1～6の目の出現回数を取得
    chi2 = [c2[1], c2[2], c2[3], c2[4], c2[5], c2[6]] # dice2に関する1～6の目の出現回数を取得
    chi_test1 = chisquare(chi1) # dice1に関するχ二乗検定
    chi_test2 = chisquare(chi2) # dice2に関するχ二乗検定
    
    if chi_test1[1] < alpha:
        count_non_random1 += 1
    if chi_test2[1] < alpha:
        count_non_random2 += 1

print("secrets.choiceを利用したときランダムでないと判定された回数：", count_non_random1)
print("random.randintを利用したときランダムでないと判定された回数：", count_non_random2)
```

```python
# 演習3(コードを追加してプログラムを完成させよ)

from sympy import gcdex
import sys

def solve_cong_equation(a, b, n): # 1次合同式 ax≡b (mod n) を解く関数
    x, y, d = sympy.gcdex(a, n) # ax+by=d を満たす x, y, d を計算
    if b % d > 0:
        sys.stderr.write('ax≡b (mod n) has no solution') # b が d で割れないとき解はない
        return None
    sol = [] # 合同式の解を格納するリスト
    a = a//d
    n = n//d
    b = b//d
    x0 = (b*x)%(n*d) # (a/d)x≡b/d (mod n/d) の一つの解を計算（a, b, n はdで割る前の入力知）
    for i in range(d): # x0を含め、合同式の全ての解を計算し sol に追加
        sol.append(x0 + i*n)

    return sol

# (1) 
a = 759797598748
b = 579273497
n = 75297349798533543
all_sol = solve_cong_equation(a, b, n)
print("{}x≡{} (mod {}) の解x={}".format(a, b, n, all_sol))
for sol in all_sol: # 正しい解が得られたか確認
    print(a*sol % n == b) # 正しい解が得られたら True と表示

# (2)
a = 5559529747936
b = 3739756
n = 7592734473947584338
all_sol = solve_cong_equation(a, b, n)
print("{}x≡{} (mod {}) の解x={}".format(a, b, n, all_sol))
for sol in all_sol: # 正しい解が得られたか確認
    print(a*sol % n == b) # 正しい解が得られたら True と表示

# (3)
a = 282814105706
b = 35258184
n = 642839066779736
all_sol = solve_cong_equation(a, b, n)
print("{}x≡{} (mod {}) の解x={}".format(a, b, n, all_sol))
for sol in all_sol: # 正しい解が得られたか確認
    print(a*sol % n == b) # 正しい解が得られたら True と表示
```
